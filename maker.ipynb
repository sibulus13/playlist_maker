```python
import sys
import spotipy
import spotipy.util as util
import os

from dotenv import load_dotenv
import pandas as pd

load_dotenv()
SPOTIPY_CLIENT_ID = os.getenv("SPOTIPY_CLIENT_ID")
SPOTIPY_CLIENT_SECRET = os.environ.get("SPOTIPY_CLIENT_SECRET")
SPOTIPY_REDIRECT_URI = os.environ.get("SPOTIPY_REDIRECT_URI")
username = 'hotshot9258'
scope = 'playlist-modify-public'

token = util.prompt_for_user_token(username, scope, client_id=SPOTIPY_CLIENT_ID, client_secret=SPOTIPY_CLIENT_SECRET, redirect_uri=SPOTIPY_REDIRECT_URI, show_dialog=True)
if token:
    sp = spotipy.Spotify(auth=token, retries=0, status_retries=0)
    results = sp.current_user_saved_tracks(limit=1)
    for item in results['items']:
        track = item['track']
        print(track['name'] + ' - ' + track['artists'][0]['name'])
        print(track['id'])        
else:
    print("Can't get token for", username)
```

```python
data = pd.read_csv('output_data.csv')
n_clusters = 7
# separate data by cluster
cluster_data = []
created_playlist_ids = []
for i in range(n_clusters):
    # Create a spotify playlist for each cluster
    playlist = sp.user_playlist_create(username, 'Cluster ' + str(i))
    print(playlist)
    playlist_id = playlist['id']
    created_playlist_ids.append(playlist_id)
    # add songs based off id to playlist
    cluster_data.append(data.loc[data['labels'] == i])
    
    for index, row in cluster_data[i].iterrows():
        sp.user_playlist_add_tracks(username, playlist_id, [row['id']])
        print(row['id'])
        print(row['title'])
        print(row['artist'])
        print(row['labels'])
        print('-----------------')
else:
    print("Can't get token for", username)
    
for i in range(n_clusters):
    print(cluster_data[i])
    print('-----------------')
```

```python
# delete created playlists
def delete_created_cluster_playlists():
    playlists = sp.user_playlists(username)
    # print playlist names
    for playlist in playlists['items']:
        print(playlist['name'])
        print(playlist['id'])
        print('-----------------')
        # if 'Cluster' in playlist name, delete it
        if 'Cluster' in playlist['name']:
            sp.user_playlist_unfollow(username, playlist['id'])
            print('deleted')
            print('-----------------')
```
